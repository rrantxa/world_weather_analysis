```python
#Import the random module
import random
```

```python
#The randint() function
random.randint(-90, 90)
```

Returns an integer between the interval, -90 and up to 89.

```python
#The random() function
random.random() #Gets a floating point decimal number between 0 and 1,
```

Returns a floating-point decimal number between 0 and 1.

```python
#Combining functions
random_number = random.randint(-90, 89) + random.random()
random_number
```

```python
#Getting random latitudes on a list
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
    
latitudes
```

```python
#The randrange() function
random.randrange(-90, 90, step = 1)
```

Returns a whole integer between the interval, -90 and 90 where the step is the difference between each number in the sequence.

```python
random.randrange(-90, 90, step = 3)
```

```python
#The uniform() function
random.uniform(-90, 90)
```

Returns a floating-point decimal number between the interval, -90 and 90.

```python
%timeit np.random.uniform(-90.000, 90.000, size = 1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x+= 1
    return latitudes

#Call the function with 1500.
%timeit latitudes(1500)
```
